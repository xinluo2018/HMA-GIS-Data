```python
import os
import warnings
warnings.filterwarnings('ignore')
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io.shapereader import Reader
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
```

```python
path_hma = 'extents/HMA/hma_bolch2019_fill.gpkg'
# path_hma = 'extents/HMA/hma_brun2017.gpkg'
# path_hma = 'extents/HMA/hma(Pan_TPlateau)_liu2022.gpkg'
path_lakes_global = 'lakes/GLWD_L1.gpkg'
path_lakes_tibet = 'lakes/lakes_main_tibet.gpkg'
```

### Note:   
Yarlung Zangbo also know as Brahmaputra;   
Nu Jiang also know as Salween;   
Chang Jiang also know as Yangtze River.


```python
plt.rcParams.update({'font.size':15})
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection=ccrs.Mercator())
ax.set_extent([65.0, 107.0, 24.0, 46.0])
ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5, zorder=1)
fea_hma = cfeature.ShapelyFeature(Reader(path_hma).geometries(),\
                            crs=ccrs.PlateCarree(), edgecolor='Black', alpha=0.8,\
                            linestyle='-', linewidth=1.5, facecolor='none')
fea_lakes_global = cfeature.ShapelyFeature(Reader(path_lakes_global).geometries(),\
                            crs=ccrs.PlateCarree(), edgecolor='none', alpha=0.8,\
                            linestyle='-', linewidth=0, facecolor='Blue')

ax.add_feature(fea_hma, zorder=2)
ax.add_feature(fea_lakes_global, zorder=2)
ax.stock_img()

```
